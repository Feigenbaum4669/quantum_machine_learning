Idea VQE
=======================
Variational Quantum Eigensolver (VQE) [1, 2] jest sztandarowym przykładem algorytmu, który można stosować np. w chemii kwantowej i nadaje się on do implementacji na komputerach kwantowych NISQ (ang. Noisy Intermediate Scale Quantum) <https://quantumcomputing.stackexchange.com/questions/1885/what-is-meant-by-noisy-intermediate-scale-quantum-nisq-technology>.
VQE jest algorytmem o charakterze hybrydowym - mamy w nim **komputer kwantowy** realizujący bardzo prosty obwód kwantowy oraz sterujący nim **klasyczny optymalizator**. Komputer kwantowy jest używany do dwóch rzeczy:
*   przygotowania stanu kwantowego (naszego ansatzu) $|\Psi(\theta)\rangle =U(\theta)|\Psi_i\rangle$ za pomocą pewnej sparametryzowanej operacji unitarnej $U(\theta)$ na stanie wejściowym $|\Psi_i\rangle$,
*   oszacowania wartości oczekowanej energii układu (np. cząsteczki) dla ansatzu $E=\langle \Psi(\theta)| H |\Psi(\theta) \rangle$ (na podstawie sprytnie zakodowanego Hamiltonianu elektronowego $H$). W tym celu komputer kwantowy wykonuje serię pomiarów kwantowych w odpowienich bazach.

Z kolei klasyczny optymalizator na podstawie wyników pomiaru wartości oczekiwanej energii $E$ (danej przez komputer kwantowy klayscznej informacji) wylicza funkcję kosztu na podstawie której aktualizuje parametry $\theta$ operacji unitarnej, która zostanie wykonan przez komputer kwantowy w kolejnym kroku. Kroki te następnie iterujemy aż do uzyskania zbieżności. Pozwala na oszacowania energii stanu podstawowego cząsteczki. Sposób działania algorytmu opiera się o zasadę wariacyjną Ritza <https://en.wikipedia.org/wiki/Ritz_method> - zasada wariacyjna gwarantuje, że nasza wartość oczekiwana $E$ jest zawsze $\geq$ od najmniejszej wartości własnej $H$ (czyli szukanej energii stanu podstawowego).

Wolny atom węgla ma strukturę elektronową $1s^2 2s^2 2p^2$ (stan podstawowy - 2 elektrony walencyjne, dublet). W cząsteczkach organicznych występuje w stanie wzbudzonym w którym orbitale $2s^2 2p^2$ ulegają tzw. hybrydyzacji i tworzą tetraedryczny orbital $sp^3$ obsadzony przez cztery elektrony walencyjne tworzące kwartet. Przykładowo w bazie minimalnej *STO-ng* reprezentujemy każdy stan bazowy (opisujący orbital walencyjny lub rdzenia) za pomocą n prymitywnych funkcji Gaussowskich - przybliżają one radialną część STO. Baza ta nie nadaje się jednak do profesjonalnych zastosowań (np. w publikacjach). My użyjemy bazy *3-21g*, która w zastosowaniu do benzenu dała znacznie lepsze wyniki niż *STO-3g*. Ponadto liczba funckji Gaussowskich używanych do konstukcji stanów bazowych orbitali walencyjncyh i rdzenia może być inna - w bazie *3-21g* każdy orbital rdzenia opisywany jest przez funkcję bazową powstałą z trzech funkcji Gaussowskich a każdy walencyjny przez dwie funkcje bazowe: jedną powstałą z dwóch a drugą z jednej funkcji Gaussowskiej.

Implementacja algorytmu VQE
------------------------------
Biblioteka Pennylane posiada specjalną klasę `~.VQECost` która jest dedykowana VQE. Musimy podać maszynę (*dev*), Hamiltonian kubitowy (*h*) oraz program kwantowy (definuje on obwód kwantowy czyli wysokopoziomoy przepis dla komputera kwantowego). Pennylane automatycznie tworzy wszystkie węzły kwantowe (QNodes) związane z delegowaniem zadań do komputera kwantowego - węzły z programami kwantowymi do preparacji anzatzu (zaprojektowany przez nas poniżej) oraz węzły z programami do estymację wartości oczekiwanej energii (pomiary wykonywane przez komputer kwantowy). W  wyniku postprocessingu powstałcyh w ten sposób klasycznych danych (wyników pomiarów) dostajemy wartość funkcji kosztu (energię). Mamy zatem jedną abstrakcyjną klasę która dla zadanych parametrów $\theta $ zwraca funkcję kosztu (energię). 

Zaczynamy od urządzenia czyli reprezentanta fizycznej maszyny - podajemy jedynie nazwę oraz liczbę przewodów (ang. wires) czyli kubitów, wyliczonych w poprzednim kroku na podstawie złożoności Hamiltonianu.



```python
dev = qml.device('default.qubit', wires=nr_qubits)
```

Celem w VQE jest takie wytrenowanie obwodu kwantowego aby przygotowywał on na wyjściu stan podstawowy zadanego Hamiltonianu. Obwód trzeba zaprojektować na tyle sprytnie aby nie był zbyt skomplikowany (dużo parametrów do optymalizacji to kosztowne uczenie) ale na tyle złożony aby był w stanie przygotować (choć w przybliżeniu) szukany stan podstawowy. Oczywiście zawsze można zastosować obwód uninwersalny (tj. mogący przygotować dowolny stan) lecz będzie on miał sporo parametrów. Heurystyka jest następująca: jedna warstwa składa się z pewnych bramek rotacji na wszystkich kubiach i następnie jakichś bramek splątujących np. CNOT czy SWAP (https://en.wikipedia.org/wiki/Quantum_logic_gate, zobacz też https://pennylane.readthedocs.io/en/latest/introduction/operations.html). Warstwy można składać dzięki czemu zwiększamy zakres możliwych operacji unitarnych realizowanych przez obwód (a zatem i stanów końcowych czyli naszych ansatzów). My użyjemy tylko jednej warstwy.

Nasz program kwantowy będzie prostym parametryzowanym obwodem - stosujemy po jednej dowolnej rotacji
$$R(\theta_1,\theta_2.\theta_3)=R_z(\theta_3)R_y(\theta_2)R_z(\theta_1)$$ na każdy kubit oraz 7 bramkek CNOT. Mamy więc łącznie 21 parametrów (kątów) do optymalizacji. Liczymy, że taka złożoność wystarczy (**warto przetestować inne architektury obwodów - jej dobór jest częścią problemu**).
Nasze bramki działają na stan początkowy $|\Psi_i\rangle=|110000\rangle$. Jest to artefakt implementacji dla cząsteczki wodoru $H_2$ (wymaga 4 kubitów), gdzie taki stan pozwala na dokładną reprezentację stanu podstawowego, który przy podpwiednim wyborze kodowania jest postaci $\alpha|1100\rangle + \beta|0011\rangle$. Dokładniej: $|1100 \rangle$ to w istocie stan Hartree-Focka i stan podstawowy wyraża się jako ten stan plus pewna wzbudzona poprawka $|0011\rangle$ (w bazie minimalnej). Zaproponowany obwód dla wodoru musi umieć wygenerować takie stany! **Warto pomyśleć jak efektywnie dobrać $|\Psi_i \rangle$ dla benzenu w naszej bazie (część problemu)**. Obejrzmy architekturę naszego obwodu kwantowego.



```python
Image(filename=PROJECT_PATH+"VQE/images/t1.png", width=600, height=400)
```

Należy zaznaczyć, że nasz wybór $|\Psi_i \rangle=|110000⟩$ oraz bramek CNOT nie jest przypadkowy. Stan początkowy jest stanem Hartree-Focka dla naszego problemu - (dwa elekrony zapełniają aktywny HOMO a LUMO są puste). Ponadto nasz obwód wydaje się sensowny: przeprowadza ten stan identycznościowo dla wyboru zerowych kątów. Dodatkowo widzimy, że np.  stan $|\Psi_i \rangle=|001100⟩$  (jeden HOMO i jeden LUMO pusty, drugi LUMO zapełniony) również jest realizowany (wystarczy rotacją zanegować $|q_3 \rangle$. Rzecz jasna interesuje nas otrzymywanie różnych superpozycji.

```python
def circuit(params, wires):
    qml.BasisState(np.array([1, 1, 0, 0, 0, 0]), wires=wires) #stan początkowy
    for i in wires:
        qml.Rot(*params[i], wires=i) #Rot dowolna rotacja (3 parametry) - po jednej bramce na każdy kubit
    qml.CNOT(wires=[2, 3])  #dokładamy jakieś bramki splątujące - tutaj CNOT
    qml.CNOT(wires=[2, 0])
    qml.CNOT(wires=[3, 1])
    qml.CNOT(wires=[4, 0])
    qml.CNOT(wires=[5, 1])
    qml.CNOT(wires=[3, 4])
    qml.CNOT(wires=[4, 5])

    
```

Teraz składamy wszystko razem w funkcję kosztu.

```python
cost_fn = qml.VQECost(circuit, h, dev)
```

Na tym poziomie zostaje już klasyczne uczenie maszynowe z klasycznym optymalizatorem "widzącym" jedynie swoją funkcję kosztu (*cost_fn*). Pozostaje zatem dobór klasycznego optymizatora. Pennylane dostarcza wiele możliwości (https://pennylane.readthedocs.io/en/stable/introduction/optimizers.html). Wybieramy wbudowany optymalizator gradientowy z krokiem 0.3. Parametry początkowe obwodu wyberamy losowo z rozkładu normalnego $N(0,\pi)$. Dzięki temu przy każdym uruchomienu będą inne i przez co zwiększamy szanse na otrzymanie zbieżności optymalizacji. **Zarówno wybór optymalizator jak i parametrów początkowych jest częścią problemu.**

```python
opt = qml.GradientDescentOptimizer(stepsize=0.3)
np.random.seed(0)
```

Ustalamy maksymalną liczbę kroków optymalizacji - $200$, potrzebną do uzyskania tolerancji zbieżności (różnicy między funkcją kosztu dla dwóch kolejnych kroków) $\sim 10^{
-6}$.



```python
max_iterations = 200
conv_tol = 1e-06

prev_energy = cost_fn(params)
E=[]
for n in range(max_iterations):
    params = opt.step(cost_fn, params)  #optymalizator wyznacza parametry dla obwodu wariacyjnego na podstawie wyliczonej funkcji kosztu
    energy = cost_fn(params) #wyliczamy wartość funkcji kosztu dla wyznaczonych parametrów - tu siedzi wszystko co kwantowe ...
    E.append(energy)
    conv = np.abs(energy - prev_energy)   #liczymy parametr zbieżności

    if n % 5 == 0:
        print('Iteracja = {:},  Energia stanu podstawowego = {:.8f} Ha,  Parametr zbieżności = {'
              ':.8f} Ha'.format(n, energy, conv))

    if conv <= conv_tol: #warunek zbieżności
        break

    prev_energy = energy  #aktualizacja energii

print('Przebieg uczenia:')
plt.scatter(list(range(len(E))),E) #rysujemy przebieg uczenia
plt.grid()
plt.xlabel('Iteracja')
plt.ylabel('Energia [Ha]')
plt.show()

print('Końcowy parametr zbieżności = {:.8f} Ha'.format(conv))
print('Końcowa wartość energii stanu podstawowego = {:.8f} Ha'.format(energy))
print('Dokładność względem wartości odniesienia: {:.8f} Ha ({:.8f} kcal/mol)'.
        format(np.abs(energy - (-229.4167733653)), np.abs(energy - (-229.4167733653))*627.503))
print()
print('Końcowe wartości parametrów obwodu = \n', params)
```
